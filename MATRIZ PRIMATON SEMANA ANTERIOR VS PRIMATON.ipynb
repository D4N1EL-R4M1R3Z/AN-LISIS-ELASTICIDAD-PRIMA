```python
import pandas as pd
import polars as pl
import numpy as np
import plotly.express as px

# Función para el cálculo matricial de la elasticidad
def calcular_elasticidad(df):
    # Ordena por descuento final (de menor a mayor)
    df = df.sort_values(by='descuento final').reset_index(drop=True)
    n = len(df)
    elasticidad_matriz = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            if i == j:
                elasticidad_matriz[i][j] = 1.0  # Diagonal principal
            elif i > j:
                # Solo calcular la parte inferior izquierda
                uv_i = df.at[i, 'unidades_vendidas']
                uv_j = df.at[j, 'unidades_vendidas']
                vc_i = df.at[i, 'venta_comercial']
                vc_j = df.at[j, 'venta_comercial']

                if uv_j != 0 and vc_j != 0:
                    elasticidad = abs(((uv_i - uv_j) / uv_j) / ((vc_i - vc_j) / vc_j))
                else:
                    elasticidad = np.nan
                
                elasticidad_matriz[i][j] = elasticidad
            else:
                elasticidad_matriz[i][j] = 0.0  # Triangular superior

    return elasticidad_matriz

# Función para definir el estilo de las tablas
def ventanas_color(valor):
    if pd.isna(valor):
        color = '#A9a9a9'
    elif valor == 1:
        color = '#87CEEB'
    elif 0.8 <= valor < 1:
        color = '#Adff2f'
    elif valor > 1:
        color = '#00FF00'
    elif 0 < valor < 0.8:
        color = '#Ffa07a'
    elif valor == 0:
        color = '#FF6347'
    else:
        color = '#Ffffff' 
    return f'background-color: {color}'

# Funcion ventana de color booleana
def ventanas_color_bool(valor):
    return 'background-color: #2ecc71; color: white;' if valor else 'background-color: #e74c3c; color: white;'

# Funcion del tamaño de las matrices
def tamaño_matrices(matrices_list):
    lineas_comunes = set.intersection(*[set(matrices.keys()) for matrices in matrices_list])
    
    for linea in lineas_comunes:
        descuentos_unicos = sorted(set.union(*[set(matrices[linea].index) for matrices in matrices_list]))
        
        for matrices in matrices_list:
            matriz_reajustada = matrices[linea].reindex(index=descuentos_unicos, columns=descuentos_unicos)
            matrices[linea] = matriz_reajustada
    
    return matrices_list

# Función para generar matrices de elasticidad (FILTRADO POR UDN y LINEA)
def generar_matrices_elasticidad(df):
    # Crear etiqueta UDN + Línea + Aumento descuento
    df['UDN LINEA'] = df.apply(lambda row: f"{row['UnidadDeNegocio']} {row['Linea']} {row['Aumento descuento']}", axis=1)

    # Incluir TODOS los descuentos, incluso los del 0%
    df = df[df['Descuento_final'].notna()].assign(
        Descuento_final=lambda x: (x['Descuento_final'] * 100).astype(int)
    )

    df = df.sort_values(by=['Descuento_final', 'Aumento descuento'], ascending=True)

    # Agrupación por línea y descuento
    df = df.groupby(['UDN LINEA', 'Descuento_final'])[['Unidades vendidas TL', 'Venta comercial TL']].sum().reset_index()
    df.columns = ['Linea', 'descuento final', 'unidades_vendidas', 'venta_comercial']

    # Ordenar por venta comercial para priorizar líneas más relevantes
    df = df.sort_values(by=['venta_comercial', 'descuento final', 'Linea'], ascending=[False, True, False])

    # Construcción de matrices por línea (clave = UDN + Línea + Aumento)
    claves_ordenadas = df['Linea'].unique()
    matrices_elasticidad = {}

    for clave in claves_ordenadas:
        linea_df = df[df['Linea'] == clave]
        if not linea_df.empty:
            # Ordenar por descuento final para asegurar consistencia
            linea_df = linea_df.sort_values(by='descuento final').reset_index(drop=True)

            # Recalcular elasticidad con lógica correcta
            n = len(linea_df)
            elasticidad_matriz = np.zeros((n, n))
            for i in range(n):
                for j in range(n):
                    if i == j:
                        elasticidad_matriz[i][j] = 1.0  # Diagonal
                    elif i > j:
                        uv_i = linea_df.at[i, 'unidades_vendidas']
                        uv_j = linea_df.at[j, 'unidades_vendidas']
                        vc_i = linea_df.at[i, 'venta_comercial']
                        vc_j = linea_df.at[j, 'venta_comercial']

                        if uv_j != 0 and vc_j != 0:
                            elasticidad = abs(((uv_i - uv_j) / uv_j) / ((vc_i - vc_j) / vc_j))
                        else:
                            elasticidad = np.nan
                        elasticidad_matriz[i][j] = elasticidad
                    else:
                        elasticidad_matriz[i][j] = 0.0  # Triangular superior

            # Index ordenado
            descuentos_ordenados = sorted(set(linea_df['descuento final'].values))

            # Crear DataFrame ordenado
            elasticidad_df = pd.DataFrame(
                elasticidad_matriz,
                index=linea_df['descuento final'].values,
                columns=linea_df['descuento final'].values
            ).loc[descuentos_ordenados, descuentos_ordenados]

            # Guardar la matriz en el diccionario
            matrices_elasticidad[clave] = elasticidad_df

    return matrices_elasticidad

# Matrices elasticidad por Sku:
def generar_matrices_elasticidad_sku(df):
    # Crear etiqueta UDN + Línea + Aumento descuento
    df['UDN LINEA'] = df.apply(lambda row: f"{row['UnidadDeNegocio']} {row['Linea']} {row['CodigoPLMSKU']} {row['Aumento descuento']}", axis=1)

    # Incluir TODOS los descuentos, incluso los del 0%
    df = df[df['Descuento_final'].notna()].assign(
        Descuento_final=lambda x: (x['Descuento_final'] * 100).astype(int)
    )

    df = df.sort_values(by=['Descuento_final', 'Aumento descuento'], ascending=True)

    # Agrupación por línea y descuento
    df = df.groupby(['UDN LINEA', 'Descuento_final'])[['Unidades vendidas TL', 'Venta comercial TL']].sum().reset_index()
    df.columns = ['Linea', 'descuento final', 'unidades_vendidas', 'venta_comercial']

    # Ordenar por venta comercial para priorizar líneas más relevantes
    df = df.sort_values(by=['venta_comercial', 'descuento final', 'Linea'], ascending=[False, True, False])

    # Construcción de matrices por línea (clave = UDN + Línea + Aumento)
    claves_ordenadas = df['Linea'].unique()
    matrices_elasticidad = {}

    for clave in claves_ordenadas:
        linea_df = df[df['Linea'] == clave]
        if not linea_df.empty:
            # Ordenar por descuento final para asegurar consistencia
            linea_df = linea_df.sort_values(by='descuento final').reset_index(drop=True)

            # Recalcular elasticidad con lógica correcta
            n = len(linea_df)
            elasticidad_matriz = np.zeros((n, n))
            for i in range(n):
                for j in range(n):
                    if i == j:
                        elasticidad_matriz[i][j] = 1.0  # Diagonal
                    elif i > j:
                        uv_i = linea_df.at[i, 'unidades_vendidas']
                        uv_j = linea_df.at[j, 'unidades_vendidas']
                        vc_i = linea_df.at[i, 'venta_comercial']
                        vc_j = linea_df.at[j, 'venta_comercial']

                        if uv_j != 0 and vc_j != 0:
                            elasticidad = abs(((uv_i - uv_j) / uv_j) / ((vc_i - vc_j) / vc_j))
                        else:
                            elasticidad = np.nan
                        elasticidad_matriz[i][j] = elasticidad
                    else:
                        elasticidad_matriz[i][j] = 0.0  # Triangular superior

            # Index ordenado
            descuentos_ordenados = sorted(set(linea_df['descuento final'].values))

            # Crear DataFrame ordenado
            elasticidad_df = pd.DataFrame(
                elasticidad_matriz,
                index=linea_df['descuento final'].values,
                columns=linea_df['descuento final'].values
            ).loc[descuentos_ordenados, descuentos_ordenados]

            # Guardar la matriz en el diccionario
            matrices_elasticidad[clave] = elasticidad_df

    return matrices_elasticidad

# Función para generar el reporte sobre líneas de interes en HTML
def generar_reporte_elasticidad_FIS_2025(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_2025_FISICAS.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 10%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
         </head>
         <body>
 
                 
        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav>


        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN 2025</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">E + P, FQ NAL, F T S</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2025, se ajustaron los precios de algunos productos entre el 15 a 50% de DCTO.:
        </p>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>27, 28 y 29 de Junio</td>
                </tr>
            </tbody>
        </table>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>15%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>40%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>50%</td>
                </tr>   
                <tr>
                    <td>20%</td>
                    <td>30%</td>
                </tr>
                <tr>
                    <td>20%</td>
                   <td>40%</td>
                </tr>      
                <tr>
                    <td>20%</td>
                    <td>50%</td>
                </tr>                                                       
            </tbody>
                <tr>
                    <td>30%</td>
                    <td>40%</td>
                </tr>                
                <tr>
                    <td>30%</td>
                    <td>50%</td>
                </tr>
                <tr>
                    <td>35%</td>
                    <td>45%</td>
                </tr>
                <tr>
                    <td>40%</td>
                    <td>50%</td>
                </tr>                                
        </table>

                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)

        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices

def generar_reporte_elasticidad_FIS_2024(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_2024_FISICAS.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 20%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
        </head>
        <body>

        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav> 
                
        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN 2024</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">ESTÁNDAR + PREMIUM, FQ NAL, F T S</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2024, se ajustaron los precios de algunos productos que tenian el 0, 20 y 30% de DCTO. de la siguiente forma:
        </p>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>28, 29 y 30 de Junio</td>
                </tr>

            </tbody>
        </table>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>20%</td>
                </tr>
                <tr>
                    <td>20%</td>
                   <td>40%</td>
                </tr>
                <tr>
                    <td>30%</td>
                     <td rowspan="2">40%</td>
                </tr>
            </tbody>
        </table>
                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)

        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        # Recorrer las matrices en el orden ya determinado por score
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices

def generar_reporte_elasticidad_OUT_2025(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_2025_OUTLET.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 10%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
        </head>
        <body>

        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav> 

        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN 2025</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">OUTLET</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2025, se ajustaron los precios de algunos productos entre el 15 a 50% de DCTO.:
        </p>
                
        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>27, 28 y 29 de Junio</td>
                </tr>

            </tbody>
        </table>
        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>15%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>40%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>50%</td>
                </tr>   
                <tr>
                    <td>20%</td>
                    <td>30%</td>
                </tr>
                <tr>
                    <td>20%</td>
                   <td>40%</td>
                </tr>      
                <tr>
                    <td>20%</td>
                    <td>50%</td>
                </tr>                                                       
            </tbody>
                <tr>
                    <td>30%</td>
                    <td>40%</td>
                </tr>                
                <tr>
                    <td>30%</td>
                    <td>50%</td>
                </tr>
                <tr>
                    <td>35%</td>
                    <td>50%</td>
                </tr>
                <tr>
                    <td>40%</td>
                    <td>50%</td>
                </tr>                                
        </table>

                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)

        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        # Recorrer las matrices en el orden ya determinado por score
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices

def generar_reporte_elasticidad_OUT_2024(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_2024_OUTLET.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 20%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
        </head>
        <body>

        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav> 
                
        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN 2024</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">OUTLET</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2024, se ajustaron los precios de algunos productos que tenian el 0, 20 y 30% de DCTO. de la siguiente forma:
        </p>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>28, 29 y 30 de Junio</td>
                </tr>

            </tbody>
        </table>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>20%</td>
                </tr>
                <tr>
                    <td>20%</td>
                   <td>40%</td>
                </tr>
                <tr>
                    <td>30%</td>
                     <td rowspan="2">40%</td>
                </tr>
            </tbody>
        </table>
                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)

        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices

def generar_reporte_elasticidad_TOPARA(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_TOPARA.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 10%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
         </head>
         <body>
 
                 
        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav> 

        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN - TOPARA 2025</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">TOPARA</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2025, se ajustaron los precios de algunos productos TOPARA del 0% aumentando a:
        </p>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>27, 28 y 29 de Junio</td>
                </tr>
            </tbody>
        </table>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>15%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>30%</td>
                </tr>                                                   
            </tbody>                                
        </table>

                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)

        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices

def generar_reporte_elasticidad_TOPARA_KEY(df):
    # Generar las matrices de elasticidad
    matrices = generar_matrices_elasticidad(df)

    # Crear el archivo HTML
    with open('Primaton vs Semana anterior_TOPARA_key.html', 'w', encoding='utf-8') as f:
        f.write("""<!DOCTYPE html>
        <html lang="es">
        <head>
            <meta charset="utf-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Matriz de Elasticidad</title>
            <style>
                body { font-family: 'Arial', sans-serif; margin: 30px; background-color: #f4f4f9; }
                h1 { color: #2c3e50; text-align: center; font-size: 36px; margin-bottom: 20px; }
                h2 { color: #2980b9; font-size: 24px; margin-bottom: 10px; }
                .color-subtitulo { color: #9b59b6; font-size: 20px; }
                .color-TL1 { color: #0D47A1; }
                .color-Prom { color: #2ecc71; }
                .color-CV { color: #8e44ad; }
                table { width: 100%; border-collapse: collapse; margin-bottom: 30px; border-radius: 8px; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1); }
                th, td { padding: 12px; text-align: center; border: 1px solid #ddd; }
                th { background-color: #3498db; color: white; font-weight: bold; }
                td { background-color: #ffffff; }
                tr:nth-child(even) td { background-color: #ecf0f1; }
                tr:nth-child(odd) td { background-color: #ffffff; }
                tr:hover td { background-color: #f1c40f; }
                /* Contenedor para las matrices */
                .matrix-row { display: flex; justify-content: flex-start; flex-wrap: wrap; gap: 20px; margin-bottom: 20px; }
                .matrix-container { flex: 1 1 10%; box-sizing: border-box; text-align: center; }
                /* Aseguramos que solo haya 5 matrices por fila máximo */
                .matrix-container:nth-child(5n+1) { margin-left: 0; }
                .matrix-title { text-align: center; margin-bottom: 10px; font-size: 20px; font-weight: bold; }
                .no-disponible { padding: 20px; background-color: #f39c12; color: white; font-size: 18px; font-weight: bold; text-align: center; }
                .explanation { background-color: #f44336; color: white; padding: 20px; border-radius: 8px; font-size: 16px; text-align: center; margin-top: 10px; }
            </style>
         </head>
         <body>
 
                 
        <style>
          nav {
            background-color: #3498db;
            padding: 10px;
            text-align: center;
            position: relative;
          }

          nav div {
            display: inline-block;
            position: relative;
          }

          nav a {
            margin: 0 15px;
            color: white;
            text-decoration: none;
            font-weight: bold;
          }

          .submenu {
            display: none;
            position: absolute;
            background-color: #2980b9;
            min-width: 200px;
            z-index: 1;
            text-align: left;
            top: 100%;
            left: 0;
            border-radius: 6px;
          }

          .submenu a {
            color: white;
            padding: 10px;
            display: block;
            text-decoration: none;
          }

          nav div:hover .submenu {
            display: block;
          }

          /* Submenú dentro del submenú */
          .submenu-item {
            position: relative;
          }

          .subsubmenu {
            display: none;
            position: absolute;
            top: 0;
            left: 100%;
            background-color: #1f618d;
            min-width: 180px;
            border-radius: 6px;
            box-shadow: 0 2px 6px rgba(0, 0, 0, 0.2);
          }

          .submenu-item:hover .subsubmenu {
            display: block;
          }

          .subsubmenu a {
            padding: 10px;
            color: white;
            text-decoration: none;
          }

          nav a:hover,
          .submenu a:hover,
          .subsubmenu a:hover {
            background-color: #1abc9c;
            color: white;
          }
        </style>

      </head>
      <body>
      
        <nav>
          <div>
            <a href="#">Primatón TOTTO 2025</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2025_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2025_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOTTO 2024</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_FISICAS.html">E+P, FQ NAL, F T S</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_FIS.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_2024_OUTLET.html">OUTLET</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_2024_OUT.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

          <div>
            <a href="#">Primatón TOPARA</a>
            <div class="submenu">
              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA.html">SIN KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA.html">Resumen</a>
                </div>
              </div>

              <div class="submenu-item">
                <a href="Primaton vs Semana anterior_TOPARA_key.html">KEYPAGO</a>
                <div class="subsubmenu">
                  <a href="Resumen_sku_TOPARA_KEY.html">Resumen</a>
                </div>
              </div>
            </div>
          </div>

        </nav> 

        <h1>MATRIZ DE ELASTICIDAD PRIMATÓN - TOPARA con KEYPAGO 2025</h1>
        <h2 class="color-subtitulo" style="text-align:center;">ELASTICIDAD POR LÍNEA</h2>
        <h2 class="color-subtitulo" style="text-align:center;">TOPARA con KEYPAGO</h2>
        <p style="text-align:center; font-size:16px; margin-bottom: 20px;">
            Para la primatón de junio de 2025, se ajustaron los precios de algunos productos TOPARA aumentando el 10% de descuento para compra con keypago así:
        </p>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>PRIMATÓN</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>27, 28 y 29 de Junio</td>
                </tr>
            </tbody>
        </table>

        <table style="margin: 0 auto; width: 50%;">
            <thead>
                <tr>
                    <th>DCTO. Previo a Primatón</th>
                    <th>DCTO. aplicado Primatón</th>
                    <th>DCTO. aplicado KEYPAGO</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>0%</td>
                    <td>15%</td>
                    <td>25%</td>
                </tr>
                <tr>
                    <td>0%</td>
                    <td>30%</td>
                    <td>40%</td>
                </tr>                                                   
            </tbody>                                
        </table>

                                
        <p style="text-align:center; font-size:16px; margin-top: 20px;">
            A continuación se presentan las matrices de elasticidad para visualizar el efecto del aumento de los descuentos por línea:
        </p>
        """)


        # Agrupar por 'UnidadDeNegocio' y 'Linea' y generar matrices
        from collections import defaultdict

        # Agrupar por "UnidadDeNegocio Línea" (sin el aumento)
        matrices_agrupadas = defaultdict(list)
        for clave in matrices:
            partes = clave.split()
            if 'DCTO.' in partes:
                idx = partes.index('DCTO.')
                udn_linea = ' '.join(partes[:idx])
                aumento = ' '.join(partes[idx:])
            else:
                udn_linea = clave
                aumento = ''

            matrices_agrupadas[udn_linea].append((aumento, matrices[clave]))

        # Recorrer por grupo de UDN + Línea
        for udn_linea, lista_matrices in matrices_agrupadas.items():
            f.write(f"<div class='matrix-title'><h2>{udn_linea}</h2></div>")
            f.write("<div class='matrix-row'>")  # fila de matrices

            for aumento_texto, matriz in lista_matrices:
                f.write("<div class='matrix-container'>")
                estilo_matriz = matriz.style.format('{:.3f}').applymap(ventanas_color).set_table_attributes('style=\"width:100%;\"').to_html()
                f.write(estilo_matriz)
                f.write("</div>")  # matrix-container

            f.write("</div><br><br>")  # fin de la fila de matrices


#Insertar el df de la primatón 
generar_reporte_elasticidad_FIS_2025(df_Prima_2025_FIS)
generar_reporte_elasticidad_FIS_2024(df_Prima_2024_FIS)
generar_reporte_elasticidad_OUT_2025(df_Prima_2025_OUT)
generar_reporte_elasticidad_OUT_2024(df_Prima_2024_OUT)
generar_reporte_elasticidad_TOPARA(TOPARA)
generar_reporte_elasticidad_TOPARA_KEY(TOPARA_KEY)


print("Reportes generados correctamente con éxito")
```
